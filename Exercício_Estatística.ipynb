```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
```

### Extraindo os Dados da planilha de exercicio na aba Exercicio 1

```python
df_exercicio_1 = pd.read_excel("/content/drive/MyDrive/Lista de Exercicios Complementares_MC.xlsx", sheet_name=0) #Exercicio 1
```

### Criando um Data Frame para armazenar os dados que preciso

```python
df_tratado = pd.DataFrame()
```

### Armazenando apenas os dados que preciso da planilha

```python
df_tratado = df_exercicio_1[['Observação','Organizado']]
df_tratado = df_tratado.rename(columns={'Observação':'Observacao','Organizado':'Vl_Renda'})
df_tratado = df_tratado.dropna()
df_tratado = df_tratado.reset_index(drop=True)
```

### Criando as condições das faixas salariais

```python
condicoes = [(df_tratado ['Vl_Renda'] >= 0) & (df_tratado ['Vl_Renda'] <= 2000),
             (df_tratado ['Vl_Renda'] >= 2001) & (df_tratado ['Vl_Renda'] <= 4000),
             (df_tratado ['Vl_Renda'] >= 4001) & (df_tratado ['Vl_Renda'] <= 6000),
             (df_tratado ['Vl_Renda'] >= 6001) & (df_tratado ['Vl_Renda'] <= 8000),
             (df_tratado ['Vl_Renda'] >= 8001) & (df_tratado ['Vl_Renda'] <= 10000),
             (df_tratado ['Vl_Renda'] >= 10001) & (df_tratado ['Vl_Renda'] <= 12000)]

faixa_renda = ['0 - 2000',
               '2001 - 4000',
               '4001 - 6000',
               '6001 - 8000',
               '8001 - 10000',
               '10001 - 12000']

df_tratado['Faixa_Renda'] = np.select(condicoes, faixa_renda, default='Não Informado')

```

## Criando Frequencia Absoluta

### Quantas Observações existem por Faixa de Renda ?

```python
quantidade_por_faixa_f_absoluta = df_tratado['Faixa_Renda'].value_counts() ### conta e agrupa quantas vezes a mesma faixa de renda se repete
quantidade_por_faixa_f_absoluta = quantidade_por_faixa_f_absoluta.sort_index() ### ordena as Faixas em ordem decrescente
quantidade_por_faixa_f_absoluta = quantidade_por_faixa_f_absoluta.reset_index() ### reset index para trazer a faixa de renda como coluna
quantidade_por_faixa_f_absoluta = quantidade_por_faixa_f_absoluta.rename(columns={'index':'Faixa_Renda','count':'Frequencia_Absoluta'}) ### renomeia a coluna count
display(quantidade_por_faixa_f_absoluta.style.hide()) #dessa forma é possivel visualizar o df sem o indice no inicio
```

### Representação Grafica

```python
fig_qtd_por_faixa_f_absoluta = px.bar(quantidade_por_faixa_f_absoluta.sort_values(by='Frequencia_Absoluta', ascending=True), #Estou ordenando com base no valor do menor para o maior
                                      x='Faixa_Renda',
                                      y='Frequencia_Absoluta',
                                      text_auto=True,
                                      title='Quantidade de Pessoas por Faixa de Renda',
                                      hover_data= ['Faixa_Renda', 'Frequencia_Absoluta'],
                                      color='Frequencia_Absoluta',
                                      labels={'Frequencia_Absoluta': 'Quantidade de Pessoas por Faixa de Renda', 'Faixa_Renda': 'Faixa de Renda'},
                                      height=800)
fig_qtd_por_faixa_f_absoluta.show()
```

## Criando Frequencia Relativa

### Quantos porcento existe de uma determinada Faixa de renda comparado com toda a base de dados ?

```python
quantidade_por_faixa_f_relativa = pd.DataFrame()
quantidade_por_faixa_f_relativa['Frequencia_Relativa'] = df_tratado['Faixa_Renda'].value_counts(normalize=True) * 100 #dessa forma estou verificando a proporcao que a faixa se repete
quantidade_por_faixa_f_relativa.reset_index(inplace=True)
quantidade_por_faixa_f_relativa
```

### Representação Grafica

```python
colors =  ["gold", "mediumturquoise", "darkorange", "lightgreen","rgba(150,205,255,288)","rgba(177,133,167,255)"]
grafico_pizza_qtd_por_faixa_f_relativa = go.Figure(data=[go.Pie(labels=quantidade_por_faixa_f_relativa['Faixa_Renda'],
                                                                values=quantidade_por_faixa_f_relativa['Frequencia_Relativa'],
                                                                hole=.5, # Aqui eu defino a espessura do grafico
                                                                pull = [0, 0, 0.2], # Aqui eu defino qual fatia que vai ficar separada
                                                                marker=dict(colors = colors, pattern=dict(shape=[".","x","+","-","+","x"])), # Aqui eu defino a cor e a espessura da borda
                                                                )

])
grafico_pizza_qtd_por_faixa_f_relativa.update_layout(title_text='Quantidade de Pessoas por Faixa de Renda',

                                                    title_x=0.5)
grafico_pizza_qtd_por_faixa_f_relativa.show()
```

## Criando Frequencia Absoluta Acumulada

### Qual a representação da quantidade acumulada por Faixa de Renda ?

```python
frequencia_absoluta_acumulada = df_tratado['Faixa_Renda'].value_counts().cumsum()
frequencia_absoluta_acumulada = frequencia_absoluta_acumulada.reset_index()
frequencia_absoluta_acumulada = frequencia_absoluta_acumulada.sort_values(by='Faixa_Renda', ascending=True)
frequencia_absoluta_acumulada = frequencia_absoluta_acumulada.reset_index(drop=True)
frequencia_absoluta_acumulada = frequencia_absoluta_acumulada.rename(columns={'count':'Qtd_Absoluta_Acumulada'})
frequencia_absoluta_acumulada
```

### Representação Grafica

```python
fig_qtd_por_faixa_f_absoluta_acumulada = px.bar(frequencia_absoluta_acumulada.sort_values(by='Qtd_Absoluta_Acumulada', ascending=True), #Estou ordenando com base no valor do menor para o maior
                                      x='Faixa_Renda',
                                      y='Qtd_Absoluta_Acumulada',
                                      text_auto=True,
                                      title='Quantidade de Pessoas por Faixa de Renda - Acumulado',
                                      hover_data= ['Faixa_Renda', 'Qtd_Absoluta_Acumulada'],
                                      color='Qtd_Absoluta_Acumulada',
                                      labels={'Qtd_Absoluta_Acumulada': 'Quantidade de Pessoas por Faixa de Renda', 'Faixa_Renda': 'Faixa de Renda'},
                                      height=800)
fig_qtd_por_faixa_f_absoluta_acumulada.show()
```

## Criando Frequencia Relativa Acumulada

```python
frequencia_relativa_acumulada = pd.DataFrame()
frequencia_relativa_acumulada['Frequencia_Relativa_Acumulada'] = (df_tratado['Faixa_Renda'].value_counts(normalize=True) * 100).cumsum() #dessa forma estou verificando a proporcao que a faixa se repete
frequencia_relativa_acumulada.reset_index(inplace=True)
frequencia_relativa_acumulada = frequencia_relativa_acumulada.sort_values(by='Faixa_Renda', ascending=True)
frequencia_relativa_acumulada
```

### Representação Grafica

```python
fig_qtd_por_faixa_f_relativa_acumulada = px.bar(frequencia_relativa_acumulada.sort_values(by='Frequencia_Relativa_Acumulada', ascending=True), #Estou ordenando com base no valor do menor para o maior
                                      x='Faixa_Renda',
                                      y='Frequencia_Relativa_Acumulada',
                                      text_auto=True,
                                      title='Quantidade de Pessoas por Faixa de Renda - Acumulado',
                                      hover_data= ['Faixa_Renda', 'Frequencia_Relativa_Acumulada'],
                                      color='Frequencia_Relativa_Acumulada',
                                      labels={'Frequencia_Relativa_Acumulada': 'Quantidade de Pessoas por Faixa de Renda - Acumulado', 'Faixa_Renda': 'Faixa de Renda'},
                                      height=800)
fig_qtd_por_faixa_f_relativa_acumulada.show()
```

### Juntando os Data Frames

```python
from functools import reduce
df_para_juntar = [frequencia_absoluta_acumulada, frequencia_relativa_acumulada, quantidade_por_faixa_f_absoluta, quantidade_por_faixa_f_relativa] # defino todos os df que preciso juntar
df_final = reduce(lambda left, right: pd.merge(left, right, on='Faixa_Renda'), df_para_juntar)
df_final
```
